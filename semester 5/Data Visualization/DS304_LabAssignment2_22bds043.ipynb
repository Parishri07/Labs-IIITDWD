Distribution of mpg

```python
plt.figure(figsize=(8, 6))
sns.histplot(df['mpg'], kde=True, bins=20)
plt.title('Distribution of MPG')
plt.xlabel('MPG')
plt.ylabel('Count')
plt.show()
```

Counts of cars by cylinder type

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='cyl', data=df)
plt.title('Counts of Cars by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Count')
plt.show()
```

Proportion of automatic vs. manual transmission

```python
transmission_counts = df['am'].value_counts()
print(transmission_counts / len(df))
```

Distribution of horsepower with outlier identification

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x='hp', data=df)
plt.title('Distribution of Horsepower')
plt.xlabel('Horsepower')
plt.show()
```

Frequency distribution of number of gears

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='gear', data=df)
plt.title('Frequency Distribution of Origin')
plt.xlabel('Origin')
plt.ylabel('Count')
plt.show()
```

What is the relationship between horsepower and weight?

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.scatterplot(x='hp', y='wt', data=df)
plt.title('Horsepower vs Weight')
plt.xlabel('Horsepower')
plt.ylabel('Weight')
plt.show()
```

How does the distribution of miles per gallon (mpg) vary across different numbers of cylinders (cyl)?

```python
sns.boxplot(x='cyl', y='mpg', data=df)
plt.title('MPG Distribution by Number of Cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('MPG')
plt.show()
```

What is the correlation matrix among numerical variables?

```python
correlation_matrix = df.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

**Visualize Datasaurus dozen**

```python
import pandas as pd

df = pd.read_csv('/content/drive/MyDrive/Classroom/DS304 Visualization (Aug - Nov 2024) DSAI/DatasaurusDozen.tsv', sep='\t')

print(df.head())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Visualize a specific dataset (e.g., 'dino')
sns.scatterplot(x='x', y='y', data=df[df['dataset'] == 'dino'])
plt.title('Dino Dataset')
plt.show()
```

```python
# To visualize all datasets, you can iterate over unique dataset names
for dataset in df['dataset'].unique():
    sns.scatterplot(x='x', y='y', data=df[df['dataset'] == dataset])
    plt.title(dataset)
    plt.show()
```

```python
for dataset in df['dataset'].unique():
  subset = df[df['dataset'] == dataset]
  print(f"Dataset: {dataset}")
  print(subset.describe())
  print("\n")
```
