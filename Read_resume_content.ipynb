```python
import re
from pdfminer.high_level import extract_pages, extract_text
```

```python
for page_layout in extract_pages("Parishri resume.pdf"):
    for element in page_layout:
        print(element)
```

```python
text = extract_text("Parishri resume.pdf")
print(text)
```

```python
import csv

with open('my_data.csv', 'w', newline='') as csvfile:
    csv_writer = csv.writer(csvfile)
    csv_writer.writerows(text)
```
